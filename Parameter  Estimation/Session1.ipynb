## **Session 1** - Parameter Esimation
تخمین پارامتر در قلب یادگیری ماشین قرار دارد و به عنوان سنگ بنایی عمل می‌کند که اکثر الگوریتم‌ها بر اساس آن ساخته می‌شوند. در هسته خود، یادگیری ماشین اساساً در مورد کشف پارامترهای بهینه برای توابعی است که می‌توانند پدیده‌های دنیای واقعی را به طور دقیق توصیف و پیش‌بینی کنند.

از الگوریتم‌های کلاسیک مانند رگرسیون خطی و بیز ساده گرفته تا شبکه‌های عصبی عمیق پیشرفته، هدف اساسی همچنان ثابت است: تخمین پارامترهایی که به بهترین وجه با داده‌های مشاهده شده مطابقت دارند و الگوهای اساسی را ثبت می‌کنند.


### Definition
درواقع تخمین پارامتر فرآیندی بنیادی در یادگیری ماشین و آمار است که بهترین مقادیر پارامتر ها را برای یک مدل استنتاج شده از مجموعه داده ها تعیین می کند.

درواقع یعنی ما با استفاده از مجموعه داده یک مدل مشخص را برای پیش بینی تعیین می کنیم. پارامتر های این مدل نامعلوم هستند. حال با استفاده از **تخمین پارامتر** بر اساس داده ها. بهترین مقدار را برای هر پارامتر تعیین می کنیم.

Formally, parameter estimation can be defined as:
>The process of using sample data to calculate a numerical value (an estimate) for an unknown population parameter.

In machine learning contexts, we can expand this definition:
>The task of finding the best set of parameters for a model that maximizes its ability to explain or predict the observed data.

---

#### Key Components
1. **Model:** A mathematical representation of the relationship between variables in our data.
2. **Parameters:** The unknown values in our model that we need to estimate.
3. **Data:** The observed samples from which we estimate the parameters.
4. **Estimation Method:** The algorithm or approach used to find the best parameter values.

---

#### The Estimation Process
1. **Collect Data:** Gather a sample of observations.
2. **Choose a Model:** In this case, we've chosen the Gaussian distribution.
3. **Select an Estimation Method:** We might use Maximum Likelihood Estimation (MLE) or Method of Moments.
4. **Estimate Parameters:** Apply the chosen method to find $\hat{\mu}$ and $\hat{\sigma}$ (the "hat" notation denotes estimates).
5. **Evaluate:** Assess how well our estimated parameters fit the data.


### Types of Parameters in Machine Learning Models

#### 1. Model Parameters
پارامتر های مدل مقادیری هستند که:
- توسط خود مدل یادگرفته می شوند.
- از روی داده های آموزشی تخمین زده می شوند.
- در زمان آموزش تغییر می کنند
- معمولا با الگوریتم هایی مانند $Gradiant Descent$ بهینه می شوند

*مثال:* رگرسیون خطی

فرضا در مدل زیر:
$$ \hat{y} = w_{1} x_{1} + w_{2} x_{2} + b $$

- $w_{1}:$ پارامتر مدل
- $w_{2}:$ پارامتر مدل
- $b:$ پارامتر مدل

مدل با دیدن داده ها یاد می گیرد که بترین مقادیر برای آنها چیست

*مثال در شبکه های عصبی:*

در یک شبکه عصبی وزن ها $(Weight)$  و بایاس ها $(Biases)$ جزو پارامتر های مدل هستند

---

#### 2. Hyperparameters
مقادیر هستند که:
- مدل آنها را مستقیما یاد نمی گیرد
- قبل از آموزش تعیین می شود
- توسط انسان یا الگوریتم های جست و جو تعیین می شوند
- روی فرآیند یادگیری تاثیر می گذارند

برای مثال در شبکه های عصبی موارد زیر هایپر پارامتر هستند:
```
learning_rate = 0.001
batch_size = 32
epochs = 100
```

---

#### 3. Latent Parameters
یا پارامتر نهفته چیزی است که:
- مستقیما مشاهده نمی شود
- اما از داده های مشاهده شده استنباط می شود
- به توضیح ساختار پنهان داده کمک می کند

مثلا در یک نمونه داده ما اسم و سن و جنسیت افراد را داریم. رده سنی آنها (کودک، نوجوان، بزرگسال و ...) مستقیما در داده ها نیامده است اما از داده های موجود قابل استنباط است.

یا در مثالی در مورد شبکه های عصبی، مدل ها معمولا می توانند لایه های نهفته را استخراج کنند. مانند لبه هاريال اشکال، اجسام و ... در یک تصویر.

---

#### 4. Fixed Parameters
مقادیری هستند که در طول فرآیند یادگیری تغییر نمی کنند.

برای مثال در مدل $y = wx + b$ اگر مقدار b را از قبل تعیین کنیم و اجازه ندهیم مدل آن را تغییر دهد درواقع یک Fixed Parameters درست کرده ایم

در یادگیری ماشین، Fixed Parameter می‌تواند شامل موارد زیر باشد:

- مقدارهایی که از قبل مشخص شده‌اند
- پارامترهایی که عمداً فریز شده‌اند
- تنظیمات یک مدل از پیش آموزش‌دیده
- پارامترهای آماری که در یک مرحله ثابت نگه داشته می‌شوند

---

#### 5. Structure Parameters
این ها پارامتر هایی هستند که ساختار یا معماری مدل را تعیین می کنند.

مثلا در شبکه های عصبی پارامتر های ساختاری می توانند شامل موارد زیر باشند:

- تعداد لایه ها
- تعداد نورون های هر لایه
- نوع لایه ها
- اتصال بین لایه ها
- نو Activation Function
- ساختار درخت تصمیم
- تعداد اجزای یک مدل Mixture

---

#### 6. Regularization Parameters
این پارامتر ها برای مدیریت و کنتزل Overfitting استفاده می شوند.

### Overview of Parameter Estimation Methods

#### 1. Maximum Likelihood Estimation (MLE)
**Core Idea:** Find the parameter values that maximize the likelihood of observing the given data
**Mathematical Formulation:**
$$\hat{\theta}_{MLE} = \arg\max_{\theta} L(\theta|X) = \arg\max_{\theta} P(X|\theta)$$

---

#### 2. Bayesian Estimation
**Core Idea:** Combine prior knowledge about parameters with observed data to obtain a posterior distribution.
**Mathematical Formulation:**
$$P(\theta|X) = \frac{P(X|\theta)P(\theta)}{P(X)}$$
- $P(\theta|X)$ is posterior
- $P(X|\theta)$ is the likelihood
- $P(\theta)$ is the prior
- $P(X)$ is the Evidence

---

#### 3.Method of Moments (MoM)
**Core Idea:** Equate sample moments to theoretical moments of the distribution to solve for parameters.
**Mathematical Formulation:**
$$E[X^r] = \frac{1}{n} \sum_{i=1}^n x_i^r \quad \text{for } r = 1, 2, ..., k$$

---

#### 4. Maximum A Posteriori (MAP)
MAP estimation can be seen as a bridge between MLE and full Bayesian estimation.
**Core Idea:** Find the mode of the posterior distribution.
**Mathematical Formulation:**
$$\hat{\theta}_{MAP} = \arg\max_{\theta} P(\theta|X) = \arg\max_{\theta} [P(X|\theta)P(\theta)]$$

در این روش علاوه بر داده های یک بتور اولیه $P(\theta)$ هم داریم. این روش بیان می کند بعد از دیدن داده ها کدام مقدار $\theta$ محتمل تر است؟

درواقع: $$MAP = Likelihood . Prior$$

---

#### Comparison and Context
Each method has its strengths and is suited to different scenarios:

- **MLE** is widely used for its efficiency and simplicity, especially with large datasets.
- **Bayesian Estimation** shines when prior knowledge is available and full uncertainty quantification is needed.
- **Method of Moments** can be valuable for quick estimates or when working with complex models where MLE is intractable.
- **MAP** offers a middle ground, incorporating prior knowledge while still providing a point estimate.

### Differences Between **Likelihood** and **Probability**
فرض می کنیم در پدیده انداختن سکه داریم:
$$ \theta = P(head)$$

حالا سه بار سکه رو پرتاب کردیم و نتیجه شده:
$$ X = \{ h , h , t \}$$

---
#### Probability
در احتمال پارامتر را داریم و میدانیم و می پرسیم:
> اگر احتمال شیر آمدن سکه $\theta = 0.5$ باشد احتمال دیدن $\{h , h, t\}$ چقدر است؟

یعنی: $$ P(X | \theta = 0.5)$$

---
#### Likelihood
حالا برعکس، نتیجه پرتاب ها را قبلا دیده ایم:
$$ X = \{h , h , t\}$$

>با توجه به این داده ها، کدام مقدار $\theta$ منطقی تر است؟

---

#### Properties of Likelihood
1. **Non-negative:** Likelihood is always non-negative.
2. **Relative:** Only relative values of likelihood matter, not absolute values.
3. **Not a Probability:** Likelihood doesn't sum or integrate to 1 over parameter space.

---

#### Likelihood Function
The likelihood function is central to MLE. It's a function of the parameters, treating the observed data as fixed:
$$L(\theta) = P(X|\theta)$$

For independent and identically distributed (i.i.d.) observations, the likelihood is the product of individual probabilities:
$$L(\theta) = \prod_{i=1}^n P(x_i|\theta)$$
