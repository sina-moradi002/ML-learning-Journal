## **Session 2** - Maximum Likelihood Estimation (MLE)
این روش درواقع یک سوال بنیادی می پرسد:

> Given our observed data, what parameter values would make this data most likely to occur?
>
> با توجه به داده‌های مشاهده‌شده، چه مقادیر پارامتری احتمال وقوع این داده‌ها را افزایش می‌دهند؟

فرمول ریاضی:
$$\hat{\theta}_{MLE} = \arg\max_{\theta} L(\theta|X) = \arg\max_{\theta} P(X|\theta)$$

Where:

 - $\hat{\theta}_{MLE}$ is the MLE estimate of the parameter(s)
 - $L(\theta|X)$ is the likelihood function
 - $X$ is the observed data
 - $P(X|\theta)$ is the probability of observing $X$ given parameters $\theta$



### Mathematical Formulation of MLE

$$L(\theta|X) = P(X|\theta)$$

For independent and identically distributed (i.i.d.) observations, the likelihood is the product of individual probabilities:
$$L(\theta|X) = \prod_{i=1}^n P(x_i|\theta)$$

---

#### Log-Likelihood Function
در ماشین لرنینگ پس از مرحله تخمین پارامتر ها اغلب نیاز است که از تابع $L(\theta|X)$ مشتق بگیریم.

در فرمول بالا این تابع به صورت ضرب عبارت ها است که مشتق گیری را در آن دشوار می کند. برای راحت تر شدن محاسبات اغلب از این تابع لگاریتم میگیریم تا به صورت جمع لگاریتم آن عبارت ها در بیاید و مشتق گیری آسان تر شود.

The log-likelihood function is defined as:
$$\ell(\theta|X) = \log L(\theta|X) = \sum_{i=1}^n \log P(x_i|\theta)$$

البته این نگرش تفاوتی در هدف تابع ایجاد نمی کند:
$$\hat{\theta}_{MLE} = \arg\max_{\theta} L(\theta|X) = \arg\max_{\theta} \ell(\theta|X)$$

#### Finding the Maximum
To find the maximum, we typically follow these steps:

1. Take the derivative of the log-likelihood with respect to each parameter:
$$\frac{\partial \ell(\theta|X)}{\partial \theta_j} = 0$$

2. Solve the resulting equation(s), known as the likelihood equations:
$$\sum_{i=1}^n \frac{\partial \log P(x_i|\theta)}{\partial \theta_j} = 0$$

3. Check the second derivative to ensure it's a maximum, not a minimum:
$$\frac{\partial^2 \ell(\theta|X)}{\partial \theta_j^2} < 0$$



### Step-by-Step Process of MLE


#### 1. Identify the Probability Distribution
اولین و مهم ترین قدم این است که بهترین توزیع احتمالاتی را برای پدیده مورد بررسی پیدا کنیم

انتخاب توزیع احتمالاتی مناسب در نهایت ما را به نتیجه مطلوب می رساند. هر توزیع احتمالاتی برای هر پدیده مناسب نمی باشد.

گاهی نیاز است چندین توزیع را برای یک پدیده تست کنیم تا به بهترین انتخاب برسیم

---

#### 2. Write the Probability Function
هر توزیع احتمالاتی یک تابع احتمال مخصوص به خود دارد که باید بر اساس آن ادامه مراحل انجام شود.

احتمال مشاهده یک نقطه داده واحد را با توجه به پارامترها بیان کنید:
$$P(x_i|\theta)$$

---

#### 3. Construct the Likelihood Function
For independent observations, the likelihood is the product of individual probabilities:$$L(\theta|X) = \prod_{i=1}^n P(x_i|\theta)$$

---

#### 4. Take the Logarithm
Convert the likelihood to log-likelihood for computational ease:$$\ell(\theta|X) = \log L(\theta|X) = \sum_{i=1}^n \log P(x_i|\theta)$$

---

#### 5. Find the Maximum
برای پیدا کردن ماکسیمم های یک تابع بر اساس قوانین ریاضیات باید مشتثق تابع را برابر صفر قرار دهیم.

سپس مشتق دوم را چک کنیم تا از جواب مطمئن شویم

در توابع چند متغیره باید مشتقات جزئی تابع را نسبت به تمام متغیر ها پیدا کنیم و برابر صفر قرار دهیم و چند معادله چند مجهول حاصل را محاسبه کنیم.

اگر پس از گرفتن مشتق جزئی عبارت برای حل کردن دشوار بود باید از **روش های تخمین استفاده** شود.

---

#### 6. Solve for the Parameters
بسته به پیچیدگی معادلات: برای موارد ساده، به صورت تحلیلی حل کنید. برای موارد پیچیده، از روش‌های عددی مانند گرادیان نزولی یا نیوتن-رافسون استفاده کنید.

---




### **Example** - Normal Distribution
1. Probability Function $$ P(x_i|\mu,\sigma^2) = \frac{1}{\sqrt{2\pi\sigma^2}} e^{-\frac{(x_i-\mu)^2}{2\sigma^2}}$$
2. Log-likelihood $$ \ell(\mu,\sigma^2|X) = -\frac{n}{2}\log(2\pi\sigma^2) - \frac{1}{2\sigma^2}\sum_{i=1}^n (x_i-\mu)^2$$
3. Derivatives $$\frac{\partial \ell}{\partial \mu} = \frac{1}{\sigma^2}\sum_{i=1}^n (x_i-\mu) = 0$$ $$ \frac{\partial \ell}{\partial \sigma^2} = -\frac{n}{2\sigma^2} + \frac{1}{2(\sigma^2)^2}\sum_{i=1}^n (x_i-\mu)^2 = 0$$
4. Solve: $$ \hat{\mu}_{MLE} = \frac{1}{n}\sum_{i=1}^n x_i $$ $$\hat{\sigma}^2_{MLE} = \frac{1}{n}\sum_{i=1}^n (x_i-\hat{\mu})^2$$
