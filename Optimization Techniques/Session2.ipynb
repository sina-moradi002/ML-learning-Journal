## **Session 2** - Gradient Descent

این مفهوم درواقع به معنی شیب است که سرعت تغییرات را در مفهوم ریاضی نشان میدهد. یا سرعت رشد یا سرعت افت.

گرادیانت به عنوان یکی از شیوه های بهینه سازی به دلایل زیر کاربر بسیار مهم و زیر بنایی در ماشین لرنینگ ایفا می کند:

1. **سازگاری:** برای پاسخگویی به نیازهای متغیر این حوزه تکامل یافته است.
2. **سادگی:** مفهوم اصلی آن شهودی است و آن را برای زبان‌آموزان و متخصصان قابل فهم می‌کند.
3. **اثربخشی:** به طور مداوم در طیف وسیعی از مسائل عملکرد خوبی دارد.

---

البته گرادیانت تنها روش نیست و روش های بسیاری وجود دارند. برای مثال:

1. Newton's Method:

    - Faster convergence for some problems
    - Uses second-order derivatives
    - Computationally expensive for high-dimensional problems
2. Quasi-Newton Methods (e.g., BFGS, L-BFGS):

    - Approximate the Hessian matrix
    - Often faster than gradient descent
    - More memory-efficient than Newton's method
3. Conjugate Gradient Method:

    - Effective for large-scale problems
    - Can be faster than gradient descent for certain types of problems
4. Evolutionary Algorithms:

    - Inspired by biological evolution
    - Can handle non-differentiable objective functions
    - Useful for global optimization
5. Simulated Annealing:

    - Inspired by annealing in metallurgy
    - Can escape local minima
    - Useful for discrete optimization problems



### Concept
گرادیان یک مفهوم اساسی در حساب دیفرانسیل و انتگرال است و نقش حیاتی در الگوریتم‌های بهینه‌سازی، به ویژه در گرادیان نزولی، ایفا می‌کند. درک گرادیان برای درک نحوه عملکرد بسیاری از الگوریتم‌های یادگیری ماشین ضروری است.

گرادیان یک تابع برداری است که شیب مماس یک تابع را در یک نقطه معین نشان می‌دهد. این تابع در جهت شدیدترین افزایش تابع قرار دارد.

طبعا برای تابعی با یک متغیر این مقدار برابر با مشتق آن تابع است

اما برای یک تابع چند مغیره برابر است با برداری از مشتقات جزئی مثلا برای $f(x_1, x_2, ..., x_n)$ برابر است با:
$$\nabla f = \left(\frac{\partial f}{\partial x_1}, \frac{\partial f}{\partial x_2}, ..., \frac{\partial f}{\partial x_n}\right)$$

ویژگی های مهم:

- **جهت:** گرادیان در جهت تندترین صعود تابع قرار دارد.

- **بزرگی:** بزرگی گرادیان نشان دهنده نرخ افزایش تابع در آن جهت است.

- **گرادیان صفر:** در یک نقطه مینیمم، ماکزیمم یا زینی محلی، گرادیان صفر است.



### Importance in machine learning
In machine learning, gradients are crucial for several reasons:

1. **Optimization:** They guide us in finding the minimum of loss functions.
2. **Feature Importance:** They help in understanding which features have the most impact on the output.
3. **Backpropagation:** They are essential in calculating how to adjust weights in neural networks.



### Mathematical Representation
for function $ f(x)$ :
$$ x_{new} = x_{old} - \alpha \nabla f(x_{old}) $$
Where:

 - $x_{new}$ is the new point
 - $x_{old}$ is the current point
 - $\alpha$ is the *learning rate* (step size)
 - $\nabla f(x_{old})$ is the gradient at the current point

درواقع در گرادیان بیان می کند که در یک تابع قدم قدم پایین تر برویم تا به نقطه مینیمم برسیم.

### Learning Rate and Its Importance $(\alpha)$

مقدار $\alpha$ درواقع نشان می دهد هر پرش به سمت پایین چقدر است.

این پرش ها نباید خیلی زیاد یا خیلی کم باشند.

اگر خیلی کم باشند روند کند می شود و حجم محاسبات ما بسیار بالا می رود. . همچنین ریسک گیر کردن در مینیمم های محلی (مینیمم نسبی) را افزایش می دهد.

اگر زیادی بزرگ باشد باعث می شود هر قدم از حد فراتر رود و از مینیمم بگذرد. از طرفی می تواند باعث نوسان و واگرایی شود. همچنین بهینگی را از بین می برد.

اما یک $\alpha$ مناسب می تواند به طور موثر به حداقل همگرا ‌شودو  سرعت و دقت را متعادل می‌کند.
