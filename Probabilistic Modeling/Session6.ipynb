## **Session 6** - Statistical Inference: Frequentist vs Bayesian Approaches
**Frequentist Approach:**
این مورد یعنی ما یک موضوع رو تنها بر اساس آنچه می بینیم و اطلاعاتی که داریم بررسی و قضاوت کنیم. درواقع از اطلاعات قبلی استفاده نمی کنیم یا حتی اصلا اطلاعات قبلی وجود ندارد و بررسی های ما صرفا بر اساس اطلاعات موجود و مشاهدات است.

**Bayesian Approach:**
این مورد علاوه بر مشاهدات کنونی، تجربه های گذشته و اطلاعات قبلی را هم در بررسی ما مورد توجه قرار می دهد. این شیوه متداول تر است.


### Frequentist Approach

**Advantages:**

- ✅ Widely accepted and taught
- ✅ Clear-cut decision rules
- ✅ Doesn't require prior knowledge


**Limitations:**

- ❌ Can be unintuitive (what exactly is a p-value?)
- ❌ Doesn't incorporate prior knowledge
- ❌ May lead to overconfidence in results

### Bayesian Approach
**Advantages:**
- ✅ Widely accepted and taught
- ✅ Clear-cut decision rules
- ✅ Doesn't require prior knowledge


**Limitations:**
- ❌ Can be unintuitive (what exactly is a p-value?)
- ❌ Doesn't incorporate prior knowledge
- ❌ May lead to overconfidence in results
