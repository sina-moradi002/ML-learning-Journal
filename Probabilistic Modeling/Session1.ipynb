### Random Variable
متغیری است که مقدار آن بسته به یک احتمال یا شانس بستگی دارد. مانند پرتاب یک تاس. درواقع در مورد مقدار آن عدم قطعیت وجود دارد.

دو نوع گسسته (Dicrete) و پیوسته (Continuous) را شامل می شود.



### Probability Distribution
یک تابع ریاضی است که مقداری که یک Random Variable میتواند به خود بگیرد را شرح می دهد.

[All Probabilistic Distributions](https://www.math.wm.edu/~leemis/chart/UDR/UDR.html)


### Joint, Marginal, and Conditional Probability
- **Marginal:** Probability of a single event EXP: P(man)
- **Joint:** Probability of two or more events EXP: P(man & engineer)
- **Conditional:** probability of an event conditional on another event EXP: P(man | engineer)

### Independence
برخی احتمالات وابستگی به هم دیگر ندارند. مثلا اگر A و B مستقل باشند:
P(A and B) = P(A) * P(B)


### Advantages
- عدم قطعیت و نویز در دیتا را هندل می کند
- دانش قبلی (Prior Knowledge) را در محاسبات دخیل بکنیم.
- نتایج آن قابل تفسیر است.

### Limitations
- محاسبات پیچیده
- نیاز به فرضیات در مورد Data Distribution. فرضیات ممکن است مشکل داشته باشند یا به کلی اشتباه باشند.
- در ابعاد بالا هندل کردن داده ها سخت می شود.
