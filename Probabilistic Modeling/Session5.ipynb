## **Session 5** - Common Continuous Probability Distributions
آن‌ها احتمالات مقادیر ممکن برای متغیرهای تصادفی پیوسته - متغیرهایی که می‌توانند هر مقداری را در یک محدوده معین بپذیرند - توصیف می‌کنند. برخلاف همتایان گسسته خود، توزیع‌های پیوسته با تعداد نامتناهی از نتایج ممکن سروکار دارند.


### Uniform Distribution
این حالت وقتی است که متغیر شما می‌تواند هر عدد حقیقی را در یک بازه (فاصله) مشخص بگیرد (نه فقط اعداد صحیح).

For example, if we want to randomly select a real number between 0 and 1, the probability that the number is between 0 and 0.1 is 0.1. The probability that the number is between 0.1 and 0.2 is also 0.1, and so on...

---

#### Probability Density Function
$$ f(x) = \begin{cases}
\frac{1}{b-a} & \text{for } a \leq x \leq b \\
0 & \text{otherwise}
\end{cases} $$

---

#### mean and variance
For $$ X~U(a,b)$$:
- mean: $$ E[X] = (a+b)/2 $$
- Variance: $$ Var(X) = (b-a)^2 / 12 $$

### Normal (Gaussian) Distribution
مهم ترین و متداول ترین توزیع آماری است.

![Normal Distribution](img.png)

We denote this as X ~ N(μ, σ²), where σ² is the variance.

---

#### Standard Normal Distribution

The Standard Normal distribution is a special case where μ = 0 and σ = 1. We denote this as Z ~ N(0, 1).

Any Normal distribution can be converted to the Standard Normal using the formula: $$ Z = (X - μ) / σ $$
This process is called standardization or z-score normalization.

---

#### PDF
$$ f(x) = \frac{1}{\sigma \sqrt{2\pi}} e^{-\frac{1}{2}(\frac{x-\mu}{\sigma})^2} $$
For the Standard Normal distribution, this simplifies to:
$$ f(z) = \frac{1}{\sqrt{2\pi}} e^{-\frac{z^2}{2}} $$

---

#### Mean and Variance
mean: $$ E[X] = μ $$
variance: $$Var(X) = σ²$$

---

#### Central Limit Theorem
اگر تعداد نمونه‌های شما به اندازه کافی بزرگ باشد، توزیع میانگین آن نمونه‌ها، بدون توجه به شکل توزیع جامعه اصلی، به توزیع نرمال (منحنی زنگی) نزدیک می‌شود.
نکته این است که این قضیه ربطی به خود داده ها ندارد و درباره **توزیع میانگین داده ها** است.


### Exponential Distribution
توزیع نمایی، توزیع «زمان انتظار» برای اتفاقاتی است که ناگهانی و مستقل از هم رخ می‌دهند، و هر چه بیشتر منتظر بمانید، احتمال اینکه اتفاق در ثانیه‌ی بعد بیفتد، کمتر و کمتر می‌شود (اما هیچ‌وقت به صفر نمی‌رسد).

**مثال:**
تصور کنید یک لامپ کم‌مصرف جدید خریداری کرده‌اید.

سال اول: احتمال اینکه لامپ بسوزد نسبتاً زیاد است.

سال سوم: اگر لامپ تا الان سالم مانده، احتمال سوختن آن در سال سوم کمتر از سال اول است (چون لامپ‌های ضعیف‌تر قبلاً سوخته‌اند و آنهایی که مانده‌اند مقاوم‌ترند).

سال دهم: احتمال اینکه این لامپ در ثانیه‌ی بعد بسوزد، بسیار بسیار ناچیز است.

اگر عمر این لامپ را روی یک نمودار بکشید، یک منحنی می‌بینید که از یک نقطهٔ بلند در سمت چپ (لحظهٔ صفر) شروع می‌شود و به صورت نمایی (یا هذلولوی) به سمت راست پایین می‌آید و به صفر میل می‌کند، اما هرگز به صفر نمی‌رسد (چون همیشه احتمال ریاضیِ بسیار ریزی برای سوختن وجود دارد).

---

این مهم‌ترین ویژگی توزیع نمایی است و آن را از بقیه توزیع‌ها جدا می‌کند:

احتمال اینکه یک رویداد در ۵ دقیقهٔ آینده رخ دهد، کاملاً مستقل از این است که شما قبلاً چقدر منتظر مانده‌اید.

---

#### PDF
$$ f(x; λ) = \begin{cases}
λe^{-λx} & \text{for } x ≥ 0 \\
0 & \text{for } x < 0
\end{cases} $$
Where:
- x is the value of the random variable (usually time)
- λ is the rate parameter
