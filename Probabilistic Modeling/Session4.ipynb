## **Session 4** - Common Discrete Probability Distribution
توزیع‌های احتمال گسسته، احتمال وقوع هر مقدار از یک متغیر تصادفی گسسته را توصیف می‌کنند.
به یاد داشته باشید، یک متغیر تصادفی گسسته فقط می‌تواند تعداد قابل شمارشی از مقادیر متمایز، مانند اعداد صحیح یا مجموعه‌ای متناهی از نتایج، را بپذیرد.


### Bernoulli Distribution
ساده ترین توزیع است که فقط شامل دو حالت است: یا موفق می شود یا شکست میخورد.

success probability = p
failure probability = 1-p

for a bernoulli random variable:
$$ P(X = x) = \begin{cases}
p & \text{if } x = 1 \text{ (success)} \\
1-p & \text{if } x = 0 \text{ (failure)}
\end{cases}$$

**mean:**
$$ E[X] = p $$
**Variance:**
$$ Var(X) = p(1-p) $$

### Binomial Distribution
یک توزیع احتمال گسسته بنیادی است که تعداد موفقیت‌ها را در تعداد ثابتی از آزمایش‌های مستقل برنولی مدل‌سازی می‌کند.

#### Relationship with bernoulli
The Binomial distribution is essentially a sum of n independent Bernoulli trials:
If X₁, X₂, ..., Xn are independent Bernoulli(p) random variables, then:
- X = X₁ + X₂ + ... + Xn ~ Bin(n, p)

#### Probability Mass Function
For a Binomial(n, p) distribution, the probability of exactly k successes is given by:
$$ P(X = k) = \binom{n}{k} p^k (1-p)^{n-k} $$

#### mean and variance
**mean:**
$$ E[X] = np $$
**variance:**
$$ Var(X) = np(1-p) $$


### Multinomial Distribution
در این حالت مقادیر ممکن برای احتمال ما میتوانند بیشتر از دو حالت داشته باشند. مثلا پرتاب یک تاس؛ که میتواند 6 حالت داشته باشد

#### Probability Mass Function
For a Multinomial(n, p₁, p₂, ..., pₖ) distribution, the probability of observing x₁ occurrences of outcome 1, x₂ occurrences of outcome 2, and so on, is given by:
$$ P(X_1 = x_1, X_2 = x_2, ..., X_k = x_k) = \frac{n!}{x_1! x_2! ... x_k!} p_1^{x_1} p_2^{x_2} ... p_k^{x_k} $$

#### Mean and Variance
**mean:**
$$ E[X_i] = np_i $$
**variance:**
$$ Var(X_i) = np_i(1-p_i) $$


### Poisson Distribution
این توزیع احتمال تعداد دفعات رخ دادن یک پدیده را در یک بازه زمانی یا مکانی خاص بررسی می کند.
برای مثال چقدر احتمال دارد در یک ساعت 20 تا مشتری از یک مغازه بازدید کنند.

Key characteristics of a Poisson distribution:

1. It deals with discrete, countable outcomes (represented by k).
2. Events occur randomly and independently.
3. The average rate of occurrence (λ) is known and constant.
4. λ is the only parameter needed to describe the distribution.

#### Probability Mass Function
$$ P(X = k) = \frac{e^{-λ} λ^k}{k!} $$
